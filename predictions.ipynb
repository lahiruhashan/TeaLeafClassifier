```python
from pathlib import Path
import cv2
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
CATEGORIES = ['ready', 'not-ready']
home_dir = str(Path.home())
def prepare(filepath):
    IMG_SIZE = 50
    img_array = cv2.imread(filepath, cv2.IMREAD_GRAYSCALE)
    new_array = cv2.resize(img_array, (IMG_SIZE, IMG_SIZE))
    return new_array.reshape(-1, IMG_SIZE, IMG_SIZE, 1)
```

```python
model = tf.keras.models.load_model(home_dir + "/PycharmProjects/TeaLeafClassifier/model/classifier.model")
```

```python
img_array = cv2.imread(home_dir + '/Documents/ready.JPG')
plt.imshow(img_array, cmap='gray')
plt.show()
prediction = model.predict([prepare(home_dir + '/Documents/ready.JPG')])
print(CATEGORIES[int(prediction[0][0])])
```

```python
img_array = cv2.imread(home_dir + '/Documents/not-ready.JPG')
plt.imshow(img_array, cmap='gray')
plt.show()
prediction = model.predict([prepare(home_dir + '/Documents/not-ready.JPG')])
print(CATEGORIES[int(prediction[0][0])])
```
