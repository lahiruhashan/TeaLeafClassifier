```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
import time
```

```python
img = cv2.imread('gc2.jpg')
```

```python
green = np.uint8([[[0, 255, 0]]]) #here insert the bgr values which you want to convert to hsv
hsvGreen = cv2.cvtColor(green, cv2.COLOR_BGR2HSV)
print(hsvGreen)

lowerLimit = hsvGreen[0][0][0] - 10, 100, 100
upperLimit = hsvGreen[0][0][0] + 10, 255, 255

print(upperLimit)
print(lowerLimit)
```

```python
hsv_img = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)
sen = 30
green1 = (60-sen,30,100)
green2 = (60+sen,255,255)
colorMask = cv2.inRange(hsv_img, green1, green2)
result = cv2.bitwise_and(img, img, mask=colorMask)

plt.imshow(result)
plt.show()
```

```python
NAME = str(time.time())
cv2.imwrite("grabcut/" + str(time.time()) +  ".jpeg",result)
```

```python
img = cv2.imread('gc2.jpg')
mask = np.zeros(img.shape[:2],np.uint8)

dimensions = img.shape

bgdModel = np.zeros((1,65),np.float64)
fgdModel = np.zeros((1,65),np.float64)

rect = (0,0,1500,1499)
cv2.grabCut(img,mask,rect,bgdModel,fgdModel,3,cv2.GC_INIT_WITH_RECT)

mask2 = np.where((mask==3)|(mask==1),0,1).astype('uint8')
img = img*mask2[:,:,np.newaxis]
plt.imshow(img),plt.colorbar(),plt.show()
```

```python
NAME = str(time.time())
cv2.imwrite("grabcut/" + str(time.time()) +  ".jpeg",img)
```
